```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))

def deriv_sigmoid(z):
    return sigmoid(z)*(1 - sigmoid(z))

def bipolar_sigmoid(z):
    return 2/(1 + np.exp(-z)) -1

def deriv_bipolar_sigmoid(z):
    return 0.5 * (1 + bipolar_sigmoid(z)) * (1 - bipolar_sigmoid(z))
```
